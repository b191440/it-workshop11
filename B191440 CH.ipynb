```python
price = list(df["Price"])
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
plt.hist(x=price, edgecolor = "black" , color = "yellow" , bins=10)
plt.xlabel("price range")
plt.ylabel("number of houses")
plt.title("histogram of price")
plt.show()
```
